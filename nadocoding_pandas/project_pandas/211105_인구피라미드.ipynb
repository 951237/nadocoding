```python
import pandas as pd
```

```python
filename = '201108_201108_연령별인구현황_월간.xlsx'
```

```python
# 파일 불러오기
df_m = pd.read_excel(filename, skiprows=3, index_col='행정기관', usecols='B, E:Y')
```

```python
# 숫자의 콤모 없애기
df_m.iloc[0] = df_m.iloc[0].str.replace(',','').astype(int)
```

```python
df_m.head()
```

```python
# 여자데이터 가져오기
df_w = pd.read_excel(filename,skiprows=3, index_col='행정기관', usecols='B, AB:AV')  # 3개 줄 빼고, 행정기관 칼럼을 인덱스, 남자 인원 칼럼만 선택
df_w.head()
```

```python
# 여자 데이터프레임 칼럼을 남자 칼럼과 통일
df_w.columns = df_m.columns
```

```python
df_w.iloc[0] = df_w.iloc[0].str.replace(',',"").astype(int)
```

```python
df_w.head()
```

## 데이터 시각화

```python
import matplotlib.pyplot as plt
import matplotlib
import matplotlib.font_manager as fm
matplotlib.rcParams['font.family'] = 'UnDotum' # 리눅스 돋움체 설정
matplotlib.rcParams['font.size'] = 15 # 폰트 크기
matplotlib.rcParams['axes.unicode_minus'] = False  # - 마이너스 글꼴 오류 처리
```

```python
df_m.columns
```

```python
plt.figure(figsize=(10,7))
plt.barh(df_m.columns, -df_m.iloc[0] // 1000) # 남자인구그래프 음의 방향으로 그리기
plt.barh(df_w.columns, df_w.iloc[0] // 1000)  # 여자인구그래프 양의 방향으로 그리기
plt.title('2011년 남여 인구피라미드')
plt.savefig('2011년 남여 인구피라미드.png', dpi=100)
plt.show()
```
