# 데이터 선택 조건

```python
import pandas as pd

data = {
    '지원번호' : ['1번','2번','3번','4번','5번','6번','7번','8번'],
    '이름' : ['채치수', '정대만', '송태섭', '서태웅', '강백호', '변덕규', '황태산', '윤대협'],
    '학교' : ['북산고', '북산고', '북산고', '북산고', '북산고', '능남고', '능남고', '능남고'],
    '키' : [197, 184, 168, 187, 188, 202, 188, 190],
    '국어' : [90, 40, 80, 40, 15, 80, 55, 100],
    '영어' : [85, 35, 75, 60, 20, 100, 65, 85],
    '수학' : [100, 50, 70, 70, 10, 95, 45, 90],
    '과학' : [95, 55, 80, 75, 35, 85, 40, 95],
    '사회' : [85, 25, 75, 80, 10, 80, 35, 95],
    'SW특기' : ['Python', 'Java', 'Javascript', '', '', 'C', 'PYTHON', 'C#']
}

df = pd.DataFrame(data)
df
```

```python
# 데이터 값을 조건에 맞게 True, False 반환
df['키'] >= 185
```

```python
# 조건에 맞는 값을 선택
filt = df['키'] >= 185
df[filt]   # 키가 185이상의 값 출력
```

```python
df[~filt]  # 키가 185 미만의 데이터 선택
```

```python
# df.loc[row_sel, col_sel]
df.loc[(df['키'] >= 185), '키']
```

## 다양한 조건

```python
# & and
df.loc[(df['키'] >= 185) & (df['학교'] == '북산고')]  # 키가 185이면서 학교가 북산고인 학생 선택
```

```python
# or |
df.loc[ (df['키'] >= 185) | (df['키'] > 190) ]
```

## str함수

```python
filt = df['이름'].str.startswith('송')  # '송'으로 시작하는 데이터 선택
df[filt]
```

```python
filt = df['이름'].str.contains('태')  # 이름 칼럼에서 '태'를 포함한 데이터 선택
df[filt]
```

```python
# isin
langs = ['Python', 'Java']
filt = df['SW특기'].isin(langs)  # langs리스트의 데이터가 있는 데이터 선택
df[filt]
```

```python
langs = ['python', 'java']
filt = df['SW특기'].str.lower().isin(langs)  # sw특기 칼럼의 값을 소문자로 바꿔서 리스트의 값이 포함되어 있는지 찾기
df[filt]
```

```python
# 결측치가 있는 경우 'Flase'처리
filt = df['SW특기'].str.contains("Java", na = True)
df[filt]
```
