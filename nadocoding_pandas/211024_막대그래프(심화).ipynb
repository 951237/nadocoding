```python
import matplotlib.pyplot as plt
import matplotlib
import matplotlib.font_manager as fm
matplotlib.rcParams['font.family'] = 'UnDotum' # 리눅스 돋움체 설정
matplotlib.rcParams['font.size'] = 15 # 폰트 크기
matplotlib.rcParams['axes.unicode_minus'] = False  # - 마이너스 글꼴 오류 처리
```

```python
label = ['강백호', '서태웅', '정대만']
values = [190, 180, 184]
```

```python
plt.barh(label, values)
plt.xlim(175, 195)
```

```python
# 그래프 무늬 주기
bar = plt.barh(label, values)
bar[0].set_hatch('/') # line texture
bar[1].set_hatch('x') # cross texture
bar[2].set_hatch('..') # dot texture
```

```python
# 그래프에 값 넣기
bar = plt.bar(label, values)
plt.ylim(175, 195)
for idx, rect in enumerate(bar):
	plt.text(idx, rect.get_height()+0.5, values[idx], ha='center')
```
