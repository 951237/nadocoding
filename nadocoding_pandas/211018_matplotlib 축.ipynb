```python
import matplotlib.pyplot as plt
import matplotlib
import matplotlib.font_manager as fm
matplotlib.rcParams['font.family'] = 'UnDotum' # 리눅스 돋움체 설정
matplotlib.rcParams['font.size'] = 15 # 폰트 크기
matplotlib.rcParams['axes.unicode_minus'] = False  # - 마이너스 글꼴 오류 처리

x = [1, 2, 3]
y = [4, 2, 6]
```

```python
plt.plot(x, y)
```

## 마커

```python
plt.plot(x, y, lw = 5) # lw = linewidth
```

```python
lst_marker = ['o', 'v', 'x']
lst_marker_title = ['마커 o', '마커 v', '마커 x']
for i in range(3):
    plt.title(lst_marker_title[i])
    plt.plot(x, y, marker = lst_marker[i])  # 마커가 점 모양
    plt.show()
```

```python
lst_linestyle = [':', '--', '-.']
lst_title = ['도트 스타일', '점선 스타일', '점선 스타일2']
for i in range(3):
    plt.title(lst_title[i])
    plt.plot(x, y, marker='o', linestyle = lst_linestyle[i])
    plt.show()
```

```python
# 선색깔 지정하기
plt.plot(x, y, color = 'pink')
```

## 그래프 색, 마커, 라인스타일 포맷

```python
lst_format = ['ro--', 'bv:', 'go']
format_title = ['포맷 1 : ro--', '포맷 2 : bv:', '포맷 3 : go']

for i in range(3):
    plt.title(format_title[i])
    plt.plot(x, y, lst_format[i])  # 선색깔 'red', 마커모양 'o', 라인스타일 '--'
    plt.show()
```

## 축약어

```python
plt.plot(x,y, marker='o', mfc = 'red', ms = 10, mec = 'blue')  # mfc = markerfacecolor, ms = markersize, mec = markeredgecolor
```

```python
plt.plot(x,y, marker='o', mfc = 'red', ms = 10, mec = 'blue', alpha=0.2)  # alpha 투명도 조절
```

## 그래프의 크기

```python
plt.figure(figsize=(10, 5))  # 그래프의 크기 가로 10, 세로 5
plt.plot(x, y)
```

```python
plt.figure(figsize=(10, 5), dpi = 100)  # 그래프 확대 dpi
plt.plot(x, y)
```

```python
plt.figure(facecolor='yellow')  # 그래프 배경정하기
plt.plot(x,y)
```
