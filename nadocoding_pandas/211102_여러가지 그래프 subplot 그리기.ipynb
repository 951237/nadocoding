# 211102 여러그래프

```python
import matplotlib.pyplot as plt
import matplotlib
import matplotlib.font_manager as fm
matplotlib.rcParams['font.family'] = 'UnDotum' # 리눅스 돋움체 설정
matplotlib.rcParams['font.size'] = 15 # 폰트 크기
matplotlib.rcParams['axes.unicode_minus'] = False  # - 마이너스 글꼴 오류 처리
```

```python
import pandas as pd
df = pd.read_csv('slamdunk.csv')
df.head()
```

```python
# 한번에 여러 그래프 그리기
fig, axs = plt.subplots(2,2, figsize=(15, 10))
fig.suptitle('여러 그래프 넣기')

# 첫번째 그래프 바그래프
axs[0,0].bar(df['이름'], df['국어'], label='국어점수')  # 바그래프
axs[0,0].set_title('첫번째 그래프')  # 제목 달기
axs[0,0].legend()  # 범례표시
axs[0,0].set(xlabel='이름', ylabel='점수')  # x축, y축 이름 표시
axs[0,0].set_facecolor('lightyellow')  # 배경색지정
axs[0,0].grid(linestyle='--', linewidth=0.5)  # 그리드 표시, 선스타일과 두깨

# 두번째 그래프 꺽은선 그래프
axs[0,1].plot(df['이름'], df['수학'], label = '수학')
axs[0,1].plot(df['이름'], df['영어'], label = '영어')
axs[0,1].legend()

# 세번째그래프 수직 바그래프
axs[1,0].barh(df['이름'], df['키'])

# 네번째 그래프 꺽은선 그래프
axs[1,1].plot(df['이름'], df['사회'], color='green', alpha=0.5)
plt.show()
```
