```python
import pandas as pd

data = {
    '지원번호' : ['1번','2번','3번','4번','5번','6번','7번','8번'],
    '이름' : ['채치수', '정대만', '송태섭', '서태웅', '강백호', '변덕규', '황태산', '윤대협'],
    '학교' : ['북산고', '북산고', '북산고', '북산고', '북산고', '능남고', '능남고', '능남고'],
    '키' : [197, 184, 168, 187, 188, 202, 188, 190],
    '국어' : [90, 40, 80, 40, 15, 80, 55, 100],
    '영어' : [85, 35, 75, 60, 20, 100, 65, 85],
    '수학' : [100, 50, 70, 70, 10, 95, 45, 90],
    '과학' : [95, 55, 80, 75, 35, 85, 40, 95],
    '사회' : [85, 25, 75, 80, 10, 80, 35, 95],
    'SW특기' : ['Python', 'Java', 'Javascript', '', '', 'C', 'PYTHON', 'C#']
}

df = pd.DataFrame(data)
df
```

```python
# 칼럼의 내용 수정
df['학교'].replace({'북산고' : '상북고'}, inplace=True)  # 북산고 → 상북고
df
```

```python
# 소문자 변환
df['SW특기'] = df['SW특기'].str.lower()
df
```

```python
df['SW특기'] = df['SW특기'].str.upper()  # 영어를 모두 대문자로 바꾸기
df
```

```python
# 칼럼 추가
df['총합'] = df['국어'] + df['수학'] + df['영어'] + df['과학'] + df['사회']
df
```

```python
# 데이터 업데이트 하기
df['결과'] = 'Fail'
df.loc[df['총합'] > 400, '결과'] = 'Pass'
df
```

```python
# 칼럼 삭제
df.drop(columns = ['국어'])  # 국어 칼럼 삭제
```

```python
# 칼럼 여러개 삭제
df.drop(columns = ['국어', '수학'])  # 국어, 수학 칼럼 삭제
```

```python
# Row 삭제
df.drop(index=4)  # 4번 인덱스 삭제
```

```python
# 조건에 맞는 행 삭제
filt = df['수학'] < 80
df[filt]  # filt에 해당하는 데이터 보여주기
df[filt].index  # filt에 해당하는 인덱스 
df.drop(index = df[filt].index)  # 조건에 해당하는 인덱스를 제거하고 데이터 보여주기
```
